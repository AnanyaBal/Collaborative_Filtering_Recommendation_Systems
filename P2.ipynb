```python

tr = np.zeros((943,1682))
for i in range(943):
    for j in range (1682):
        tr[i,j]=data[i,j]
te=np.zeros((943,1682))

for i in range(943):
    big=0
    a=[]
    for j in range (1682):
        if(tr[i,j]>big):
            big=tr[i,j]
    for j in range(1682):
        if(tr[i,j]==big):
            a.append(j)
    l=len(a)
    #print(a)
    for k in range(1):
        r=random.randint(0,l-1)
        print(r,end=' ')
        val=a[r]
        print(val)
        te[i,val]=tr[i,val]
        tr[i,val]=0

            
np.savetxt('train.csv', tr, delimiter=',', fmt='%10.40f')
np.savetxt('test.csv', te, delimiter=',', fmt='%10.40f')
```

```python
#normalisation of the rating matrix
norm=np.zeros((943,1683))
for i in range(943):
    sum=0
    c=0
    for j in range(1682):
        norm[i,j]=tr[i,j]
        if(tr[i,j]!=0):
            sum+=tr[i,j]
            c+=1
    norm[i,1682]=(sum/c)
for i in range(943):
    for j in range(1682):
        if(norm[i,j]!=0):
            norm[i,j]=norm[i,j]-norm[i,1682]
        
np.savetxt('norm.csv', norm, delimiter=',', fmt='%10.40f')
```

```python
#similarity calculations w/o normalisation
import math
cossim=np.zeros((1682,1682))
              
for i in range(1682):
    for j in range(i,1682):
        num=0
        d1=0
        d2=0
        for k in range(943):
            num+=tr[k,i]*tr[k,j]
            d1+=math.pow(tr[k,i],2)
            d2+=math.pow(tr[k,j],2)
        s=num/((math.sqrt(d1))*(math.sqrt(d2)))
        cossim[i,j]=s
        cossim[j,i]=s
    print(i)
print(cossim)
np.savetxt('cossim.csv', cossim, delimiter=',', fmt='%10.40f')
```

```python
#similarity calculations w normalisation
import math
ncossim=np.zeros((1682,1682))
              
for i in range(1682):
    for j in range(i,1682):
        num=0
        d1=0
        d2=0
        for k in range(943):
            num+=norm[k,i]*norm[k,j]
            d1+=math.pow(norm[k,i],2)
            d2+=math.pow(norm[k,j],2)
        s=num/((math.sqrt(d1))*(math.sqrt(d2)))
        ncossim[i,j]=s
        ncossim[j,i]=s
    print(i)
print(ncossim)
np.savetxt('ncossim.csv', ncossim, delimiter=',', fmt='%10.40f')
```

```python
#cosine similarity neighbourhood calc w/o normalisation
n1=np.zeros((1682,10))
t=np.zeros((1682,1682))
for i in range(1682):
    for j in range(1682):
        t[i,j]=cossim[i,j]
        
for k in range(10):
    for i in range(1682):
        big=0
        val=0
        for j in range(1682):
            if(t[i,j]>big and i!=j):
                big=t[i,j]
                val=j
        n1[i,k]=(val)
        t[i,val]=-1000
np.savetxt('n1.csv', n1, delimiter=',', fmt='%10.40f')

#cosine similarity neighbourhood calc w normalisation
n2=np.zeros((1682,10))
t=np.zeros((1682,1682))
for i in range(1682):
    for j in range(1682):
        t[i,j]=ncossim[i,j]
        
for k in range(10):
    for i in range(1682):
        big=0
        val=0
        for j in range(1682):
            if(t[i,j]>big and i!=j):
                big=t[i,j]
                val=j
        n2[i,k]=(val)
        t[i,val]=-1000
np.savetxt('n2.csv', n2, delimiter=',', fmt='%10.40f')
```

```python
#Probability similarity calc w/o normalisation
ps=np.zeros((1682,1682))

for i in range(1682):
    for j in range(1682):
        num=0
        f1=0
        f2=0
        for k in range(943):
            if(tr[k,i]!=0):
                num+=tr[k,j]
                f1+=1
            if(tr[k,j]!=0):
                f2+=1
        ps[i,j]=num/(f1*(math.pow(f2,0.5)))
    print(i)
print(ps)
np.savetxt('probsim.csv', ps, delimiter=',', fmt='%10.40f')
```

```python
#Probability calc w normalisation
nps=np.zeros((1682,1682))

for i in range(1682):
    for j in range(1682):
        num=0
        f1=0
        f2=0
        for k in range(943):
            if(norm[k,i]!=0):
                num+=norm[k,j]
                f1+=1
            if(norm[k,j]!=0):
                f2+=1
        if(f1*(math.pow(f2,0.5))!=0):
            nps[i,j]=num/(f1*(math.pow(f2,0.5)))
        else:
            nps[i,j]=0
    print(i)
print(nps)
```

```python
#probability similarity neighbourhood calc w/o normalisation
n3=np.zeros((1682,10))
t=np.zeros((1682,1682))
for i in range(1682):
    for j in range(1682):
        t[i,j]=ps[i,j]
        
for k in range(10):
    for i in range(1682):
        big=0
        val=0
        for j in range(1682):
            if(t[i,j]>big and i!=j):
                big=t[i,j]
                val=j
        n3[i,k]=(val)
        t[i,val]=-1000
np.savetxt('n3.csv', n3, delimiter=',', fmt='%10.40f')

#probability similarity neighbourhood calc w normalisation
n4=np.zeros((1682,10))
t=np.zeros((1682,1682))
for i in range(1682):
    for j in range(1682):
        t[i,j]=nps[i,j]
        
for k in range(10):
    for i in range(1682):
        big=0
        val=0
        for j in range(1682):
            if(t[i,j]>big and i!=j):
                big=t[i,j]
                val=j
        n4[i,k]=(val)
        t[i,val]=-1000
np.savetxt('n4.csv', n4, delimiter=',', fmt='%10.40f')
```

```python
#Item-based Top_N Collaborative filtering

U=np.zeros((943,1682))
for i in range(943):
    c=0
    for j in range(1682):
        if(tr[i,j]!=0):
            U[i,c]=(j+1)
            c+=1

np.savetxt('U.csv', U, delimiter=',', fmt='%10.40f')

```

```python

#print(k)
C=np.zeros((943,1682))
for i in range(943):
    for j in range(1682):
        C[i,j]=-1

for i in range(943):
    Un=[]
    for j in range(1682):
        if(U[i,j]!=0):
            val=U[i,j]
            for m in range(10):
                Un.append(n1[int(val-1),m])
    print(i)
    l=len(Un)
    c=0
    for n in range(l):
        ar=C[i]
        if(Un[n] not in ar):
            C[i,c]=Un[n]
            c+=1
        
print(C)
np.savetxt('C.csv', C, delimiter=',', fmt='%10.40f')
```

```python
#Calculating the final C
for i in range(943):
    for j in range(1682):
        val=C[i,j]
        for k in range(1682):
            v=U[i,k]
            if(val==v):
                C[i,j]=-1
    print(i)
print(C)
np.savetxt('Cf.csv', C, delimiter=',', fmt='%10.40f')
```

```python
#Finding similarity between C and U with cosine similarity
S=np.zeros((943,1682))

for i in range(943):
    c=C[i]
    u=U[i]
    for j in range(1682):
        sim=[]
        val=c[j]
        sums=0
        if(val>=0):
            for k in range(1682):
                if((u[k]-1)>=0):
                    v=int(u[k])-1
                    s=cossim[int(val),int(v)]
                    sim.append(s)
            sim=sorted(sim, key=abs, reverse=True)
            #print(len(sim),end=' ')
            for l in range(10):
                sums+=sim[l]
            S[i,j]=sums
    print(i)

np.savetxt('Csim.csv', S, delimiter=',', fmt='%10.40f')
```

```python
#Finding the top N items 
tpn=np.zeros((943,10))

Scpy=np.zeros((943,1682))
for i in range(943):
    for j in range(1682):
        Scpy[i,j]=S[i,j]


for i in range(943):
    for k in range(10):
        big=0
        val=0
        for j in range(1682):
            if(Scpy[i,j]>big):
                big=Scpy[i,j]
                val=j
        tpn[i,k]=C[i,val]
        Scpy[i,val]=-1
np.savetxt('topn.csv', tpn, delimiter=',', fmt='%10.40f')
```

```python
#Recall calculation
rec1=0
res=0
for i in range(943):
    for j in range(10):
        val=tpn[i,j]
        if(te[i,int(val)]!=0):
            res+=1
rec1=res/943
print(rec1)
```
